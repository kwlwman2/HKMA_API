# Code to download Economic Data

```python
import requests 

url = 'https://api.hkma.gov.hk/public/market-data-and-statistics/monthly-statistical-bulletin/financial/economic-statistics'

r = requests.get(url)

json_data = r.json()
```

```python
# results is the key, records is also the keys, now we can get the real-time economic data via json format
data = json_data['result']['records']

headers = list(data[0].keys())

dataList = []

for i in data:
    lista = list(i.values())
    dataList.append(lista)
    
import pandas as pd

df = pd.DataFrame(dataList, columns = headers)

df 
```

# Code to download Capital Market Data


```python
import pandas as pd
import requests


url_cm = "https://api.hkma.gov.hk/public/market-data-and-statistics/monthly-statistical-bulletin/financial/capital-market-statistics"

r2 = requests.get(url_cm)

json_data_2 = r2.json()

# results is the key, records is also the keys, now we can get the real-time economic data via json format
data_cm = json_data_2['result']['records']

headers = list(data_cm[0].keys())

dataList = []

for i in data_cm:
    lista = list(i.values())
    dataList.append(lista)
    
import pandas as pd

df = pd.DataFrame(dataList, columns = headers)

df 
```
